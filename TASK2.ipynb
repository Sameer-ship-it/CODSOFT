```python
# CREDIT CARD FRAUD DETECTION

'''
Build a model to detect fraudulent credit card transactions. Use a
dataset containing information about credit card transactions, and
experiment with algorithms like Logistic Regression, Decision Trees,
or Random Forests to classify transactions as fraudulent or
legitimate.
'''

#Importing Train Data

import pandas as pd

train_df = pd.read_csv(r"C:\Users\shima\Downloads\Compressed\archive_3\fraudTrain.csv")
test_df = pd.read_csv(r"C:\Users\shima\Downloads\Compressed\archive_3\fraudTest.csv")
print("The train data is :-\n")
display(train_df)
print("The test data is :\n")
display(test_df)
train_df.info()
```

```python
# Dropping unnecessary columns of the data frame
import pandas as pd

train_df.drop(["Unnamed: 0", 'cc_num', 'first', 'last', 'street', 'trans_num', 'dob', 'trans_date_trans_time'], axis=1, inplace=True)
test_df.drop(["Unnamed: 0", 'cc_num', 'first', 'last', 'street', 'trans_num', 'dob', 'trans_date_trans_time'], axis=1, inplace=True)
display(train_df)
display(test_df)
```

```python
# Spliting & Encoding with Target Encoder

import category_encoders as ce

X_train = train_df.drop(['is_fraud'] , axis = 1)
y_train= train_df['is_fraud']

X_test = test_df.drop(['is_fraud'] , axis = 1)
y_test = test_df['is_fraud']

#Applying Target Encoding

target_cols = ['merchant', 'city', 'job']
encoded = ce.TargetEncoder(cols=target_cols, smoothing=0.2)
encoded.fit(X_train, y_train)

X_train_encoded = encoded.transform(X_train)
X_test_encoded = encoded.transform(X_test)


print("Training Encoded Data :\n")
display(X_train_encoded.head(3))
print("Test Encoded Data : \n")
display(X_test_encoded.head(3))

```

```python
#Enoding with One Hot Encoder

ohe_cols = ['category', 'gender', 'state']
X_train_ohe = pd.get_dummies(X_train_encoded[ohe_cols], drop_first=True)
X_test_ohe = pd.get_dummies(X_test_encoded[ohe_cols], drop_first=True)

# Aligning the columns
X_train_ohe, X_test_ohe = X_train_ohe.align(X_test_ohe, join='left', axis=1, fill_value=0)
```

```python
# Combine categorical + target encoded + numerical
target_cols = ['merchant', 'city', 'job']
num_cols = ['amt', 'zip', 'lat', 'long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long']

X_train_encodor = pd.concat([X_train_encoded[num_cols].reset_index(drop=True), X_train_encoded[target_cols].reset_index(drop=True), X_train_ohe.reset_index(drop=True)], axis=1)

X_test_encodor = pd.concat([X_test_encoded[num_cols].reset_index(drop=True), X_test_encoded[target_cols].reset_index(drop=True), X_test_ohe.reset_index(drop=True)], axis=1)
```

```python
#Standarisation of the data

from sklearn.preprocessing import StandardScaler

Sd = StandardScaler()


X_train_encodor[num_cols] = Sd.fit_transform(X_train_encodor[num_cols])
X_test_encodor[num_cols] = Sd.transform(X_test_encodor[num_cols])

print("Scaled Inputs of Training Data :- \n")
display(X_train_encodor.head(3))

print("scaled Input data of Test Data :- \n")
display(X_test_encodor.head(3))
```

```python
#Implementing Logistic Regression 

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , precision_score ,confusion_matrix , f1_score , recall_score

le = LogisticRegression(max_iter=1000, class_weight='balanced')

le.fit(X_train_encodor , y_train)

y_predict = le.predict(X_test_encodor)

Accuracy = accuracy_score(y_test , y_predict)
Precision = precision_score(y_test, y_predict)
confusion = confusion_matrix(y_test , y_predict)
f1 = f1_score(y_test , y_predict)
recall = recall_score(y_test , y_predict)

print(f"Accuracy of the model is : {Accuracy * 100:.2f}% ")  
print("Precision of the model is : ",Precision)
print("Confusion Matrix of the model is :\n" , confusion)
print("\nF1 Score of the model is :",f1)          #ideal >0.70
print("Recall Score of the model is :" , recall)  #Recall Ideal if  > 0.75



```

```python
from sklearn.ensemble import RandomForestClassifier

Random = RandomForestClassifier( n_estimators=100, n_jobs=-1,class_weight='balanced',random_state=42)

Random.fit(X_train_encodor , y_train)
y_predict = Random.predict(X_test_encodor)

Accuracy = accuracy_score(y_test , y_predict)
Precision = precision_score(y_test, y_predict)
confusion = confusion_matrix(y_test , y_predict)
f1 = f1_score(y_test , y_predict)
recall = recall_score(y_test , y_predict)

print(f"Accuracy of the model is : {Accuracy * 100:.2f}% ")  
print("Precision of the model is : ",Precision)
print("Confusion Matrix of the model is :\n" , confusion)
print("\nF1 Score of the model is :",f1)          #ideal >0.70
print("Recall Score of the model is :" , recall)  #Recall Ideal if  > 0.75


```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, f1_score, recall_score

DT = DecisionTreeClassifier(class_weight='balanced',random_state=42)

DT.fit(X_train_encodor, y_train)

y_predict = DT.predict(X_test_encodor)

Accuracy = accuracy_score(y_test, y_predict)
Precision = precision_score(y_test, y_predict)
confusion = confusion_matrix(y_test, y_predict)
f1 = f1_score(y_test, y_predict)
recall = recall_score(y_test, y_predict)

print(f"Accuracy of the model is : {Accuracy * 100:.2f}% ")
print("Precision of the model is : ", Precision)
print("Confusion Matrix of the model is :\n", confusion)
print("\nF1 Score of the model is :", f1)
print("Recall Score of the model is :", recall)
```
